Plotting sin linea.
Para trazar solo marcadores, puedes utilizar notacion de la cadena de acceso directo 'o', que significa "anillos".
 Ejeplo:Dibuja dos puntos en el diagrama, uno en la posicion (1,3) y otro en la posicion de (8,10)

```python
import matplotlib.pyplot as plt
import numpy as np

xpuntos=np.array([1,8])
ypuntos=np.array([3,10])

plt.plot(xpuntos, ypuntos, 'o')
plt.show()
```

Diagramas de dispersion(scatter)
Se emplean para mostrar las relaciones entre dos variables, detectando correlaciones o agrupaciones,
Ejemplo:

```python
import matplotlib.pyplot as plt

altura=[150, 160, 170, 180, 190, 200]
peso=[50, 60, 65, 80, 85, 95]

plt.scatter(altura, peso, color= "red", marker= "x")
plt.title("Relacion entre la altura y el peso")
plt.xlabel("Altura (cms)")
plt.ylabel("Peso(kg)")
plt.show()
```

Graficos de barras (bar y barch)
sirve para comparar categorias o variables tanto en tu orientacion vertical como horizontal
Ejemplo de barras verticales:

```python
import matplotlib.pyplot as plt

categorias=["A", "B", "C", "D"]
valores=[4, 7, 1, 9]

plt.bar(categorias, valores, color="orange")
plt.title("Grafica de barras")
plt.xlabel("Categorias")
plt.ylabel("Valores")
plt.show()
```
